# Queue size distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma


a, scale = 3, 12
size = 10000 
data = gamma.rvs(a=a, scale=scale, size=size)


plt.figure(figsize=(8, 5))
plt.hist(data, bins=50, density=True, alpha=0.6, color='b', label="Histogram")

x = np.linspace(0, max(data), 1000)
pdf = gamma.pdf(x, a=a, scale=scale)
plt.plot(x, pdf, 'r-', label="Gamma PDF")

plt.xlabel("Value")
plt.ylabel("Density")
plt.title(f"Gamma Distribution (a={a}, scale={scale})")
plt.legend()
plt.grid(True)
plt.show()
```


# Order Size distribution

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


alpha = 2.5  
x_min = 1.5/2.5 


samples = stats.pareto.rvs(alpha, scale=x_min, size=100000)


plt.figure(figsize=(8, 5))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='b', label="Histogram")


x = np.linspace(x_min, np.max(samples), 1000)
pdf = stats.pareto.pdf(x, alpha, scale=x_min)
plt.plot(x, pdf, 'r-', label="Pareto (Power-law) PDF")


plt.yscale('log')
plt.xlim(-1, 60)
plt.xlabel("Value")
plt.ylabel("Density (log scale)")
plt.title(f"Pareto Power Law Distribution (α={alpha}, x_min={x_min})")
plt.legend()
plt.grid()
plt.show()
```
