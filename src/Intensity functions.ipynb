```python
a1, b1, c1 = 64.459, 0.103, 4.932
a2, b2, c2 = 53.086, 0.110, 5.331
a3, b3, c3 = 4.900, 0.122, 0.145
a5, b5, c5 = 4.363, 0.502, 0.122

def l1(x):
    if x < 4:
        return 15 + x*18/4
    if x < 8:
        return 33
    return a1 * np.exp(-b1 * x) + c1
def l2(x):
    if x < 4:
        return 7 + 4*x
    if x < 9:
        return 26
    return a2 * np.exp(-b2 * x) + c2
def l3(x):
    return max(a3 * np.exp(-b3 * x) + c3, 0)
def l4(x):
    if x < 2.5:
        return (-1.5*x + 4)
    else:
        return (max(-0.08*x + 0.5, 0))
def l5(x):
    return (max(a5 * np.exp(-b5 * x) + c5, 0))
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 2, figsize=(20, 7))

ax = axs[0]
ax.plot(x, y, 'r.-', label='Historical Data', linewidth=2, markersize=8)
x_fit = np.linspace(min(x), max(x), 10000)
y_fit = list(map(l4, x_fit))
ax.plot(x_fit, y_fit, 'b-', label=f'Fitted Function')
ax.set_xlim(0, 25)
ax.set_ylim(0, 3.5) 
ax.grid(True, linestyle='--', alpha=0.7)
ax.set_xlabel('X(AES)')
ax.set_ylabel('Y(s^-1)')
ax.set_title('Cancel All')
ax.legend()

ax2 = axs[1]
ax2.plot(x2, y2, 'r.-', label='Historical Data', linewidth=2, markersize=8)
x2_fit = np.linspace(min(x2), max(x2), 1000)
y2_fit = list(map(l5, x2_fit))
ax2.plot(x2_fit, y2_fit, 'b-', label=f'Fitted Function')
ax2.set_xlim(0, 25)
ax2.set_ylim(0, 3.1) 
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_xlabel('X(AES)')
ax2.set_ylabel('Y(s^-1)')
ax2.set_title('Market All')
ax2.legend()

plt.show()
```
