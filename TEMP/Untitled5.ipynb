```python
def compute_stats(df):
    stats = {}
    for level in range(1, NUM_LEVELS + 1):
        level_df = df[df['level'] == level]
        stats[level] = {
            '#L': len(level_df[level_df['type'] == 'L']),
            '#C': len(level_df[level_df['type'] == 'C']),
            '#M': len(level_df[level_df['type'] == 'M']),
            'AES': level_df['size'].mean(),
            'AIT': np.diff(level_df['time']).mean() * 1000 if len(level_df) > 1 else np.nan
        }
    
    unique_times = np.unique(df['time'])
    price_changes = np.diff(df['mid_price'].reindex(unique_times, method='ffill'))
    annualized_volatility = price_changes.std() * np.sqrt(252 * 32400)
    
    stats_df = pd.DataFrame(stats).T
    stats_df.loc['Volatility'] = np.nan
    stats_df.at['Volatility', 'Volatility'] = annualized_volatility
    
    return stats_df
```

```python
def simulate_qr():
    bid_queues, ask_queues, bid_prices, ask_prices = initialize_lob()
    mid_price = INITIAL_MID_PRICE
    events = []
    t = 0
    prev_events = []
    lambda_entries = []
    lambdas = []
    count = 0
    
    while t < TOTAL_TIME:
    #while t == 0:
        lambda_scale = 1.0
        for time, boost in INTRA_DAY_BOOST.items():
            if t >= time:
                lambda_scale = boost
        
        xqueues = [bid_queues, ask_queues]
        level_lambdas = [sum(level_lambda(queue, i)) for queue in xqueues for i in range(5)]
        global_lambda = sum(level_lambdas)
        level_weights = level_lambdas/global_lambda
        side, level = level_dict[np.random.choice(range(NUM_LEVELS*2), p=level_weights)]
        
        queues = bid_queues if side == 'bid' else ask_queues
        prices = bid_prices if side == 'bid' else ask_prices
        
        q_before = queues[level]
        #q = q_before
        #lambda1, lambda2, lambda3 = l1(q), l2(q), l3(q)
        
        event_weights = level_lambda(queues, level)
        event_prob = np.array(event_weights) / sum(event_weights)
        if level == 0:
            event_type = np.random.choice(['L', 'C', 'M'], p=event_prob)
            #lambdas.append(1)
        else:
            event_type = np.random.choice(['L', 'C'], p=event_prob)
        
        #lambda_global_adjusted = LAMBDA_GLOBAL * lambda_scale
        #lambda_global_final = lambda_global_adjusted*global_lambda(bid_queues, ask_queues)
        #lambdas.append(lambda_global_final)
        lambda_global_final = global_lambda * lambda_scale
        
        #print(level_lambdas)
        #print(global_lambda)
        #print(level_weights)
        #print(side, level, event_type, size)

        dt = expon.rvs(scale=1/lambda_global_final)
        t += dt

        lambda_entries.append({
            'time': t,
            'lambda': lambda_global_final
        })

        if t > TOTAL_TIME:
            break
         
        size = stats.pareto.rvs(2.5, scale=1.5/2.5)
        #size = max(1, min(30, size))
        

        if event_type == 'L':
            queues[level] += size
        elif event_type != 'L' and q_before > 0:
            cancel_size = min(size, q_before)
            queues[level] -= cancel_size
            if queues[0] == 0:
                opp_queues = ask_queues if side == 'bid' else bid_queues
                if np.random.random() < 0.7:
                    queues.pop(0)
                    queues.append(gamma.rvs(a=2, scale=10))
                    opp_queues.insert(0,0)
                    opp_queues.pop()
                    if side == 'ask':
                        mid_price = mid_price + 0.5 * TICK_SIZE
                    else:
                        mid_price = mid_price - 0.5 * TICK_SIZE
                    if side == 'ask':
                        ask_prices = [mid_price + i * TICK_SIZE for i in range(1, NUM_LEVELS + 1)]
                    else:
                        bid_prices = [mid_price - i * TICK_SIZE for i in range(1, NUM_LEVELS + 1)]
                else:
                    #queues[0] = gamma.rvs(a=2, scale=10)
                    pass
                
        elif event_type == 'M':  
            cancel_size = min(size, q_before)
            queues[level] -= cancel_size
            if queues[0] == 0:
                #count += 1
                opp_queues = ask_queues if side == 'bid' else bid_queues
                if np.random.random() < 0.7:
                    queues.pop(0)
                    queues.append(gamma.rvs(a=3, scale=12))
                    opp_queues.insert(0,0)
                    opp_queues.pop()
                    if side == 'ask':
                        mid_price = mid_price + 0.5 * TICK_SIZE
                    else:
                        mid_price = mid_price - 0.5 * TICK_SIZE
                    if side == 'ask':
                        ask_prices = [mid_price + i * TICK_SIZE for i in range(1, NUM_LEVELS + 1)]
                    else:
                        bid_prices = [mid_price - i * TICK_SIZE for i in range(1, NUM_LEVELS + 1)]
                else:
                    #gamma.rvs(a=3, scale=12)
                    pass
                            
        events.append({
            'time': t,
            'type': event_type,
            'side': side,
            'level': level + 1,
            'size': size,
            'queue_before': q_before,
            'mid_price': mid_price
        })
        prev_events.append((t, event_type))
        if len(prev_events) > 100:
            prev_events.pop(0)

    lambda_entries = pd.DataFrame(lambda_entries)

    return pd.DataFrame(events), bid_queues, ask_queues, bid_prices, ask_prices, lambda_entries, prev_events, count
```
